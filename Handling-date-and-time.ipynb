## Working with Dates

```python
#  converting to datetime datatype
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

#### 1.Extract Year

```python
date["date-year"] = date["date"].dt.year
date["date-month"] = date["date"].dt.month
date["date-month_name"] = date["date"].dt.month_name()
date["date-date"] = date["date"].dt.day
date["date-day_of_week"] = date["date"].dt.day_of_week
date["date-day"] = date["date"].dt.day_name()
```
